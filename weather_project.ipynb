### City Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df_city = pd.read_csv('city_data.csv')
df_city
```

```python
df_city.dropna(subset = ['avg_temp'], inplace=True)
df_city
```

```python
# 5-years Moving Average

for i in range(0, df_city.shape[0]-4):
    df_city.loc[df_city.index[i+4],'10-years Moving Average'] = (df_city.iloc[i,3] + df_city.iloc[i+1,3] + df_city.iloc[i+2,3] + df_city.iloc[i+3,3] + df_city.iloc[i+4,3])/5

df_city
```

```python
# 10-years Moving Average

for i in range(0, df_city.shape[0]-9):
    df_city.loc[df_city.index[i+9],'10-years Moving Average'] = (df_city.iloc[i,3] + df_city.iloc[i+1,3] + df_city.iloc[i+2,3] + df_city.iloc[i+3,3] + df_city.iloc[i+4,3] + df_city.iloc[i+5,3] + df_city.iloc[i+6,3] + df_city.iloc[i+7,3] + df_city.iloc[i+8,3] + df_city.iloc[i+9,3])/10

df_city
```

```python
# 20-years Moving Average

for i in range(0, df_city.shape[0]-19):
    df_city.loc[df_city.index[i+19],'10-years Moving Average'] = (df_city.iloc[i,3] + df_city.iloc[i+1,3] + df_city.iloc[i+2,3] + df_city.iloc[i+3,3] + df_city.iloc[i+4,3] + df_city.iloc[i+5,3] + df_city.iloc[i+6,3] + df_city.iloc[i+7,3] + df_city.iloc[i+8,3] + df_city.iloc[i+9,3] + df_city.iloc[i+10,3]+ df_city.iloc[i+11,3] + df_city.iloc[i+12,3] + df_city.iloc[i+13,3] + df_city.iloc[i+14,3] + df_city.iloc[i+15,3] + df_city.iloc[i+16,3] + df_city.iloc[i+17,3] + df_city.iloc[i+18,3] + df_city.iloc[i+19,3])/20

df_city
```

```python
# k-years Moving Average

k = 50

for i in range(0, df_city.shape[0] - (k - 1)):
    sum_k = sum([df_city.iloc[i + k_i, 3] for k_i in range(k)]) / k
    df_city.loc[df_city.index[i + k - 1], str(f'{k}-years Moving Average')] = sum_k

df_city
```

```python
df_city.dropna(subset = [str(f'{k}-years Moving Average')], inplace=True)
df_city
```

```python
plt.figure(figsize=(13,8))
plt.plot(df_city[['year']], df_city[[str(f'{k}-years Moving Average')]], 'r')  # str(f'{k}-years Moving Average')
plt.grid()
plt.title('Moscow Temperature Changes', fontsize=25)
plt.title(str(f'{k} years'), loc='left')
plt.xlabel('Year', fontsize=18)
plt.ylabel('Moving Average Temperature', fontsize=18)
plt.tick_params(labelsize=13)
# plt.show()
plt.savefig('city_'+str(f'{k}years')+'.png');
```

# PROJECT IN ONE CELL

```python
for k in [10, 20, 30, 40, 50]:
    
    #  city
    df_city = pd.read_csv('city_data.csv')
    df_city.dropna(subset = ['avg_temp'], inplace=True)
    
    for i in range(0, df_city.shape[0] - (k - 1)):
        sum_k = sum([df_city.iloc[i + k_i, 3] for k_i in range(k)]) / k
        df_city.loc[df_city.index[i + k - 1], str(f'{k}-years Moving Average')] = sum_k

    df_city.dropna(subset = [str(f'{k}-years Moving Average')], inplace=True)

    #  moscow plot
    plt.figure(figsize=(13,8))
    plt.plot(df_city[['year']], df_city[[str(f'{k}-years Moving Average')]], 'r')  # str(f'{k}-years Moving Average')
    plt.grid()
    plt.title('Moscow Temperature Changes', fontsize=25)
    plt.title(str(f'{k} years'), loc='left')
    plt.xlabel('Year', fontsize=18)
    plt.ylabel('Moving Average Temperature', fontsize=18)
    plt.tick_params(labelsize=13)
    plt.savefig('city_'+str(f'{k}years')+'.png');
    
    #  global
    df_global = pd.read_csv('global_data.csv')

    df_global.dropna(subset = ['avg_temp'], inplace=True)

    for i in range(0, df_global.shape[0] - (k - 1)):
        sum_k = sum([df_global.iloc[i + k_i, 1] for k_i in range(k)]) / k
        df_global.loc[df_global.index[i + k - 1], str(f'{k}-years Moving Average')] = sum_k
    
    df_global.dropna(subset = [str(f'{k}-years Moving Average')], inplace=True)

    #  global plot
    plt.figure(figsize=(13,8))
    plt.plot(df_global[['year']], df_global[[str(f'{k}-years Moving Average')]], 'g')  # str(f'{k}-years Moving Average')
    plt.grid()
    plt.title('Global Temperature Changes', fontsize=25)
    plt.title(str(f'{k} years'), loc='left')
    plt.xlabel('Year', fontsize=18)
    plt.ylabel('Moving Average Temperature', fontsize=18)
    plt.tick_params(labelsize=13)
    plt.savefig('global_'+str(f'{k}years')+'.png');
    
    #  comparing
    fig, axes = plt.subplots(figsize=(13,8))

    axes.plot(df_city[['year']], df_city[[str(f'{k}-years Moving Average')]], label='Moscow Moving Average Temperature', color='r')
    axes.plot(df_global[['year']], df_global[[str(f'{k}-years Moving Average')]], label='Global Moving Average Temperature', color='g')

    axes.set_title('Global and Moscow Temperature Changes', fontsize=25)
    axes.set_xlabel('Year', fontsize=18)
    axes.set_ylabel('Moving Average Temperature', fontsize=18)
    axes.legend(fontsize=15)
    axes.set_title(str(f'{k} years'), loc='left')
    axes.grid(linestyle='-', linewidth=1)
    axes.tick_params(labelsize=13)
    fig.savefig(str(f'{k}years')+'.png');
```
